```python
import os
import re

def match_transcripts_and_videos(transcripts_dir, videos_dir):
    matched_pairs = []
    
    # 1. Walk through the root of the transcripts folder
    for root, dirs, files in os.walk(transcripts_dir):
        for transcript_file in files:
            if not transcript_file.endswith('.txt'):
                continue
                
            # Get the relative folder path (e.g., "grade01/subject01" or "grade_01\subject_01")
            rel_path = os.path.relpath(root, transcripts_dir)
            
            # 2. Extract the date from the end of the transcript filename
            # Looks for exactly 8 digits right before the .txt extension
            t_match = re.search(r'(\d{8})\.txt$', transcript_file)
            if not t_match:
                continue
                
            date_str = t_match.group(1) # e.g., "01092020"
            t_day = date_str[:2]        # "01"
            t_month = date_str[2:4]     # "09"
            
            # 3. Locate the corresponding folder in the videos directory
            video_folder = os.path.join(videos_dir, rel_path)
            if not os.path.isdir(video_folder):
                continue
                
            # 4. Search through the videos in that corresponding folder
            for video_file in os.listdir(video_folder):
                if not video_file.endswith('.mp4'):
                    continue
                    
                # Find the first two groups of numbers separated by an underscore
                # This works for "01_02_..." as well as "resized_cropped_01_02_..."
                v_match = re.search(r'(\d{2})_(\d{2})', video_file)
                if not v_match:
                    continue
                    
                v_day = v_match.group(1)
                v_month = v_match.group(2)
                
                # 5. Compare the day and month
                if t_day == v_day and t_month == v_month:
                    matched_pairs.append({
                        'folder': rel_path,
                        'transcript': transcript_file,
                        'video': video_file
                    })
                    
    return matched_pairs
```

```python
TRANSCRIPTS_ROOT = r"D:\KRSL_OnlineSchool\KRSL_OnlineSchool_final_dataset\transcripts\elarna_transcripts_sorted_2020_2021"
VIDEOS_ROOT = r'D:\KRSL_OnlineSchool\KRSL_OnlineSchool_final_dataset\video\elarna_video_sorted_2020_2021' 

matches = match_transcripts_and_videos(TRANSCRIPTS_ROOT, VIDEOS_ROOT)

print(f"Found {len(matches)} matches!\n")
for match in matches:
    print(f"Folder: {match['folder']}")
    print(f"  Transcript: {match['transcript']}")
    print(f"  Video:      {match['video']}\n")
```
